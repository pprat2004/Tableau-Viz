```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

# -----------------------------
# Sample Data
# -----------------------------
data = [12, 15, 14, 10, 18, 20, 22, 19, 17, 13]
series = pd.Series(data)

# -----------------------------
# Calculate Skewness & Kurtosis
# -----------------------------
skew_value = skew(series)
kurt_value = kurtosis(series)     # Fischer Kurtosis (Normal dist = 0)
pearson_kurt = kurtosis(series, fisher=False)  # Pearson Kurtosis (Normal dist = 3)

print("Data:", data)
print("Skewness:", skew_value)
print("Kurtosis (Fisher):", kurt_value)
print("Kurtosis (Pearson):", pearson_kurt)

# -----------------------------
# Histogram + KDE Curve
# -----------------------------
plt.figure(figsize=(10, 5))
plt.hist(series, bins=5, density=True)
plt.title("Histogram with Density Curve")
plt.xlabel("Values")
plt.ylabel("Frequency")

# KDE plot
series.plot(kind='kde')

plt.show()

# -----------------------------
# Boxplot for visualizing skewness
# -----------------------------
plt.figure(figsize=(6, 5))
plt.boxplot(series, vert=True)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# -----------------------------
# Interpretation printed
# -----------------------------
print("\nINTERPRETATION:")

# Skewness
if skew_value > 0:
    print(" The data is POSITIVELY skewed (tail on the right).")
elif skew_value < 0:
    print(" The data is NEGATIVELY skewed (tail on the left).")
else:
    print(" The distribution is symmetric.")

# Kurtosis
if kurt_value > 0:
    print(" The data is LEPTOKURTIC (heavy tails, sharp peak).")
elif kurt_value < 0:
    print(" The data is PLATYKURTIC (light tails, flatter peak).")
else:
    print(" The data is MESOKURTIC (normal-like).")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
import pandas as pd

# ------------------------------------------------
# Create THREE distributions
# ------------------------------------------------

# Right-skewed (long tail on the right)
right_skewed = np.random.exponential(scale=2, size=1000)

# Left-skewed (mirror exponential)
left_skewed = -np.random.exponential(scale=2, size=1000)

# High kurtosis (leptokurtic  sharp peak, heavy tails)
kurt_data = np.concatenate([np.random.normal(0, 1, 900), 
                            np.random.normal(0, 0.2, 100)])

# ------------------------------------------------
# Function to display stats + graph
# ------------------------------------------------
def show_distribution(data, title):
    data_series = pd.Series(data)

    print(f"\n{title}")
    print("Skewness:", skew(data_series))
    print("Kurtosis:", kurtosis(data_series))

    plt.figure(figsize=(8, 4))
    plt.hist(data, bins=30, density=True)
    data_series.plot(kind="kde")
    plt.title(title)
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.show()

# ------------------------------------------------
# Display all 3 distributions
# ------------------------------------------------

show_distribution(right_skewed, "Right Skewed Distribution")
show_distribution(left_skewed, "Left Skewed Distribution")
show_distribution(kurt_data, "High Kurtosis Distribution")
```
